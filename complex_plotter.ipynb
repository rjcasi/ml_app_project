```python
import matplotlib.pyplot as plt

def plot_complex_numbers(z1: complex, z2: complex):
    z_sum = z1 + z2
    z_prod = z1 * z2

    points = {
        "z1": (z1.real, z1.imag),
        "z2": (z2.real, z2.imag),
        "z1 + z2": (z_sum.real, z_sum.imag),
        "z1 · z2": (z_prod.real, z_prod.imag),
    }

    colors = {"z1":"blue","z2":"orange","z1 + z2":"green","z1 · z2":"red"}

    fig, ax = plt.subplots(figsize=(6,6))
    ax.axhline(0,color="black")
    ax.axvline(0,color="black")
    ax.set_aspect("equal")

    for label,(x,y) in points.items():
        ax.quiver(0,0,x,y,angles="xy",scale_units="xy",scale=1,color=colors[label])
        ax.text(x,y,f"{label} ({x:.2f},{y:.2f}i)",color=colors[label])

    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title("Complex Vectors")
    plt.show()

# Example run
plot_complex_numbers(2+1j, -0.5+2j)
```
